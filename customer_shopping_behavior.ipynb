```python
import pandas as pd
```

```python
df = pd.read_csv("customer_shopping_behavior.csv")
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ','_')
```

```python
df = df.rename(columns = {'purchase_amount_(usd)' :'purchase_amount'})
```

```python
df.columns
```

```python
# create a age group
```

```python
labels = ['Young Adult','Adult','Middle-Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels)
```

```python
df['frequency_of_purchases'].head()
```

```python
#create purchase_frequensy_days
```

```python
frequency_mapping = {
   'Fortnightly': 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}
df['purchase_frequense_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
